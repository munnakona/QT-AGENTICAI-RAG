```python
from langchain.chat_models import init_chat_model
```

```python
llm = init_chat_model("gemini-2.5-flash-lite", model_provider="google_vertexai")
```

```python
from langchain_core.prompts import PromptTemplate

template = """You are an experienced cook.

List the ingredients to cook the following food item {food}
"""

prompt_template = PromptTemplate.from_template(template=template)
```

```python
# Create a chain 

chain = prompt_template | llm
```

```python
# execute the chain

response = chain.invoke({"food": "pasta"})
print(response.content)
```

```python
# execute the chain

response = chain.invoke({"food": "Biriyani"})
print(response.content)
```

```python

chain.stream({"food": "pasta"})
```

```python

chain.ainvoke({"food": "pasta"})
```

```python
txt_loader = UnstructuredLoader(file_path="/Users/munna/VScode/QT-AgenticAI/RAG/hello-langchain/notebooks/data/Pandas.pdf")
```

```python
docs = txt_loader.load()
```

```python
print(len(docs))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=20)
chunks = splitter.split_documents(docs)

print(f"chunks {len(chunks)}")
```

```python
for chunk in chunks:
    print(chunk.metadata)
```

```python
from langchain_google_vertexai import VertexAIEmbeddings
```

```python

embedding = VertexAIEmbeddings(
    model_name="text-embedding-005"    
)
```

```python
embedding
```

```python
vectors = embedding.embed_documents(["How are you?", "What are you doing?"])
```

```python

for vector in vectors:
    print(vector)
```

```python

len(vectors[0])
```

```python

embedding.embed_query("How are you?")
```

```python

from langchain_community.vectorstores import FAISS
# pass the chunks and embedding
BATCH_SIZE = 200  # must be <= 250

vector_store = None

for i in range(0, len(chunks), BATCH_SIZE):
    batch = chunks[i:i + BATCH_SIZE]

    if vector_store is None:
        vector_store = FAISS.from_documents(batch, embedding)
    else:
        vector_store.add_documents(batch)
```

```python
retriever = vector_store.as_retriever()
```

```python

results = retriever.invoke("what is pandas library used for ?", k=2)
```

```python
for result in results:
    print(result.page_content)
```
