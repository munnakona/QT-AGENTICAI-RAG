```python
from langchain.chat_models import init_chat_model 
```

```python
llm = init_chat_model("gemini-2.5-flash-lite", model_provider="google_vertexai")
```

```python
from langchain_core.prompts import PromptTemplate

template = """You are an experienced cook.

List the ingredients to cook the following food item {food}
"""

prompt_template = PromptTemplate.from_template(template=template)
```

```python
# Create a chain 

chain = prompt_template | llm
```

```python
# execute the chain

response = chain.invoke({"food": "pasta"})
print(response.content)
```

```python
# execute the chain

response = chain.invoke({"food": "Biriyani"})
print(response.content)
```

```python

chain.stream({"food": "pasta"})
```

```python

chain.ainvoke({"food": "pasta"})
```

```python
from langchain_google_vertexai import ChatVertexAI

model = ChatVertexAI(
    model_name="gemini-2.5-flash-lite",
    temperature=0.2,
    max_output_tokens=100,
)
```

```python

response = model.invoke("what is capital of france ?")
```

```python
response
```

```python
response.pretty_print()
```

```python
from langchain.messages import SystemMessage, HumanMessage

response = model.invoke([
    SystemMessage("You are an expert cook"),
    HumanMessage("How to cook sambar ?")
])
```

```python
response 
```

```python
response.pretty_print()
```

```python
response = model.invoke([
    SystemMessage("You are an expert in comic movies of 2025"),
    HumanMessage("What was the best comedy movie  in 2025 ?")
])  
response.pretty_print()
```

```python
response = model.invoke([
    SystemMessage("You are an expert in crime novels"),
    HumanMessage("What was the best crime novel of 2025?")
])  
response.pretty_print()
```
