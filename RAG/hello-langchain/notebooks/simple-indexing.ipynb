```python
# indexing
# 1. loader
from langchain_unstructured import UnstructuredLoader
```

```python
# 2. splitter
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
# 3. embeddings
from langchain_google_vertexai.embeddings import VertexAIEmbeddings
```

```python
# 4. vectorstore
from langchain_community.vectorstores import FAISS
```

```python

document_loader = UnstructuredLoader("/Users/munna/VScode/QT-AgenticAI/RAG/hello-langchain/notebooks/data/policies.txt")
```

```python

documents = document_loader.load()
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20,
    separators=["\n\n", "\n", " ", ""])
```

```python

chunks = splitter.split_documents(documents)
```

```python
# embedding

embedding = VertexAIEmbeddings(model_name="text-embedding-005")
```

```python

vector_store = FAISS.from_documents(chunks, embedding)
```

```python

retriever = vector_store.as_retriever()
```

```python

results = retriever.invoke("What are working hours")
```

```python

for result in results:
    print(result.page_content)
```
