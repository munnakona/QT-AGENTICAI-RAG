```python
# indexing
# 1. loader
from langchain_unstructured import UnstructuredLoader
from langchain_community.document_loaders import DirectoryLoader
# 2. splitter
from langchain_text_splitters import RecursiveCharacterTextSplitter
# 3. embedding  
from langchain_google_vertexai.embeddings import VertexAIEmbeddings
# 4. vector store
from langchain_community.vectorstores import FAISS
# 5. retriever
```

```python
directory_loader = DirectoryLoader(path="/Users/munna/VScode/QT-AgenticAI/RAG/hello-langchain/notebooks/docs", glob="**/*.txt",loader_cls=UnstructuredLoader)
documents = directory_loader.load()
```

```python
spillter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20,
    separators=["\n\n", "\n", " ", ""])
texts = spillter.split_documents(documents) 
```

```python


# embedding

embedding = VertexAIEmbeddings(model_name="text-embedding-005")
```

```python
vector_store = FAISS.from_documents(texts, embedding)
```

```python

retriever = vector_store.as_retriever()
```

```python
results = retriever.invoke("What is in the docs ?", k=2)
```

```python
for result in results:
    print(result.page_content)
```
